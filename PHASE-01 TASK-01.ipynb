```python
#Import all required Libraries
import requests
import pandas as pd
```

```python
#Call API and data through API
api_url = "https://api.binance.com/api/v3/ticker/24hr"
response = requests.get(api_url)
```

```python
#Put If-else statement on called API
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    df.to_csv("task1.csv", index=False)  # This will create a CSV file named "task.csv" in the current directory
else:
    print("API request to task1 failed")
```
